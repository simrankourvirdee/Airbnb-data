## Task 2b: Data Cleaning (Python)

- Check for missing values in the dataframe and display the count in ascending order. **If the values are missing, impute the values as per the datatype of the columns.**
- Check whether there are any duplicate values in the dataframe and, if present, remove them.
- Display the total number of records in the dataframe before and after removing the duplicates.

```python
## Check for missing values in the dataframe and display the count in ascending order.
missing_values=df_dropped.isnull().sum().sort_values(ascending=True)
missing_values

```

```python
for column in df_dropped.columns:
    if df_dropped[column].dtype == 'object':  # Impute missing values with an empty string for object/string columns
        df_dropped[column].fillna('', inplace=True)
    else:  # Impute missing values with the mean for numeric columns
        df_dropped[column].fillna(df_dropped[column].mean(), inplace=True)

df_dropped.head(5)
```

```python
before_removal=len(df_dropped)
before_removal
```

```python
## Check whether there are any duplicate values in the dataframe and if present remove them.
df_dropped=df_dropped.drop_duplicates()
df_dropped
```

```python
## Display the total number of records in the dataframe after removing the duplicates.
after_removal=len(df_dropped)
after_removal
```

## Task 3: Data Transformation (Any Tool)


- Rename the column `availability 365` to `days_booked`
- Convert all column names to lowercase and replace the spaces in the column names with an underscore "_".
- Remove the dollar sign and comma from the columns `price` and `service_fee`. If necessary, convert these two columns to the appropriate data type.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
## Rename the column.
df_dropped=df_dropped.rename(columns={'availability 365':'days_booked'})
df_dropped.head()
```

```python
## Convert all column names to lowercase and replace the spaces with an underscore "_"
df_dropped.columns=df_dropped.columns.str.lower()
df_dropped.columns=df_dropped.columns.str.replace(' ','_')
df_dropped.head(5)
```

```python
## Remove the dollar sign and comma from the columns. If necessary, convert these two columns to the appropriate data type.
df_dropped.service_fee=df_dropped.service_fee.str.replace('$', '')
df_dropped.price=df_dropped.price.str.replace('$', '')
df_dropped.service_fee=df_dropped.service_fee.str.replace(',', '')
df_dropped.price=df_dropped.price.str.replace(',', '')
df_dropped.head()
```

## Task 5a: Data Visualization (Any Tool)

- Create a horizontal bar chart to display the top 10 most expensive neighborhoods in the dataset.
    - Create another chart with the 10 cheapest neighborhoods in the dataset.
- Create a box and whisker chart that showcases the price distribution of all listings split by room type.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
# Sort the dataframe by price in descending order
import matplotlib.pyplot as plt
sort_price=df_dropped.sort_values('price',ascending=False)
sort_price
```

```python
# Top 10 most expensive neighbourhoods
top_10 = sort_price.head(10)
top_10
```

```python
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.barplot(x='price', y='neighbourhood', data=top_10)

plt.xlabel('Price ($)')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Neighbourhood')
plt.yticks(fontsize=10)

# Add Title
plt.title('Top 10 Most Expensive Neighbourhoods')

# Add gridlines
plt.grid(axis='x', linestyle='--')

# Add data labels to the bars
for index, value in enumerate(top_10['price']):
    plt.text(value, index, f'${value:.2f}', va='center')

# Display Chart
plt.tight_layout()
plt.show()
```

```python
# Sort service fee by ascending from cheapest to most expensive
sorted_price_asc = df_dropped.sort_values('price', ascending=True)

# Bottom 10 Cheapest neighbourhoods
bottom_10 = sorted_price_asc.head(10)
bottom_10
```

```python
# Display horizontal bar chart using seaborn
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.barplot(x='price', y='neighbourhood', data=bottom_10)

plt.xlabel('Price ($)')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Neighbourhood')
plt.yticks(fontsize=10)

# Add Title
plt.title('Top 10 Cheapest Expensive Neighbourhoods')

# Add gridlines
plt.grid(axis='x', linestyle='--')

# Add data labels to the bars
for index, value in enumerate(bottom_10['price']):
    plt.text(value, index, f'${value:.2f}', va='center')

# Display Chart
plt.tight_layout()
plt.show()
```

```python
# Create box and whisker chart to show price distribution
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df_dropped)
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.title('Price Distribution by Room Type')

# Add gridlines
plt.grid(axis='y', linestyle='--')

# Show the chart
plt.tight_layout()
plt.show()
```
